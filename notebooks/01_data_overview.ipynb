### Initial Observations

- Dataset is window-based processed ECG data.
- HRV, stress, and fatigue metrics are only computed when ECG signal quality is sufficient.
- A large portion of data (~75%) has missing HRV-related values.
- Error messages indicate signal filtering failures (likely due to noise or motion).
- Stress and fatigue are not independent measurements but derived from HRV features.

### Detailed Exploration Summary

- Data is generated using sliding windows (1-minute step, ~6-minute window).
- Overlapping windows are used to improve HRV stability and temporal resolution.
- ECG signal quality strongly affects availability of HRV and derived metrics.
- Movement appears to be categorized into discrete levels (1–5), with additional continuous statistics.
- HRV-related features have high missing rates (~75%), indicating many unusable segments.
